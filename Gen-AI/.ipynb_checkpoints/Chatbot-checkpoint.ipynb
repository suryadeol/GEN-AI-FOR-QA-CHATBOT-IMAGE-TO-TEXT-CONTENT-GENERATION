```python
import matplotlib.pyplot as plt

# Sample chat history
chat_history = [
    {'user': 'User', 'message': 'Hello!', 'response': 'Hi there!', 'sentiment': 0.8},
    {'user': 'User', 'message': 'How are you?', 'response': 'I am doing well, thanks!', 'sentiment': 0.7},
    {'user': 'User', 'message': 'What can you do?', 'response': 'I can assist you with various tasks.', 'sentiment': 0.6}
]

# Extracting user messages, bot responses, and sentiment scores
user_messages = [msg['message'] for msg in chat_history]
bot_responses = [msg['response'] for msg in chat_history]
sentiment_scores = [msg['sentiment'] for msg in chat_history]

# Plotting sentiment scores over the course of the conversation
plt.figure(figsize=(10, 6))
plt.plot(sentiment_scores, marker='o', linestyle='-')
plt.title('Sentiment Analysis of Chatbot Responses')
plt.xlabel('Message Number')
plt.ylabel('Sentiment Score')
plt.xticks(range(len(chat_history)), [f'Message {i+1}' for i in range(len(chat_history))])
plt.grid(True)
plt.show()

# Creating a bar chart to visualize the length of user messages and bot responses
plt.figure(figsize=(10, 6))
plt.bar(range(len(chat_history)), [len(msg) for msg in user_messages], color='skyblue', label='User Messages')
plt.bar(range(len(chat_history)), [len(msg) for msg in bot_responses], color='orange', label='Bot Responses', bottom=[len(msg) for msg in user_messages])
plt.title('Length of User Messages and Bot Responses')
plt.xlabel('Message Number')
plt.ylabel('Length')
plt.xticks(range(len(chat_history)), [f'Message {i+1}' for i in range(len(chat_history))])
plt.legend()
plt.show()
```
